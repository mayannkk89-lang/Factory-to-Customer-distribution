```python
import pandas as pd
df = pd.read_excel("C:\\Users\\LENOVO\\Documents\\candy distribution sales.xlsx")
df.head()
```

```python
x = df[['Ship_Mode', 'Division', 'Region', 'Product_Name', 'Sales', 'Units', 'Gross_Profit', 'Cost', 'Shipping_Days']]
x.head(2)
```

```python
x_new = x.copy()
x_new.head()
```

```python
x_new['Shipping_Days'].unique()	
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
x_new['Ship_Mode'] = le.fit_transform(x_new['Ship_Mode'])
x_new['Division'] = le.fit_transform(x_new['Division'])
x_new['Region'] = le.fit_transform(x_new['Region'])
x_new['Product_Name'] = le.fit_transform(x_new['Product_Name'])
x_new.head(2)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_new['Shipping_Days'] = scaler.fit_transform(x_new[['Shipping_Days']])
x_new.head(2)
```

```python
x_new.head(1)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.model_selection import train_test_split

a = x_new[['Ship_Mode',	'Division',	'Region', 'Product_Name', 'Units', 'Shipping_Days']]
b = x_new[['Sales', 'Gross_Profit', 'Cost']]

a_train, a_test, b_train, b_test = train_test_split(a, b, test_size=0.2, random_state=42)

model1 = LinearRegression()
model1.fit(a_train, b_train)

model2 = DecisionTreeRegressor()
model2.fit(a_train, b_train)

model3 = RandomForestRegressor()
model3.fit(a_train, b_train)

model4 = KNeighborsRegressor()
model4.fit(a_train, b_train)

b_pre = model2.predict(a_test)


print("Mean Squared Error =", mean_squared_error(b_test, b_pre))
print("Mean Absolute Error =", mean_absolute_error(b_test, b_pre))
print("R² Score =", r2_score(b_test, b_pre))
```
